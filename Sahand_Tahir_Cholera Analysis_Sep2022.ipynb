<h2 style="font-family:Times New Roman Bold">3.3 Limitations </h2><br>

It is possible that these factors reduce the strength of the correlation between the parameters and the Cholera cases.<br>

<h4 style="font-family:Times New Roman Bold">4.1.1.1 Defining the Functions</h4><br>
In essence, there are two main functions required: one to search and return a list of indicators with their respective codes, and the other to take the indicator code and return the associated data in the form of a Pandas dataframe. The two functions I will use are 'find_indicator', which accepts either a key word or a list of key words and returns a dataframe containing the indicator code for those that match. The second is 'return_indicator', which returns a dataframe containing the values associated with an indicator code.

```python
#here I define the function needed to return the data from WHO GHO API
#Unit Testing is condcuted in the next block
#I define the functions below stricktly following the WHO GHO API guidelines.
#I need two functions, one to return the avaialble indicators and one to return the data for each chosen indicator

#first I start by defining a function to return the base url needed to do the request
#since the url and the appended parameters are different hence we build a function to help choose each for each purpose
def link_generator(action): 
    if not isinstance(action, str): #if the input if not string to return None
        return None
    
    action = action.lower() #to ensure capitalization does not affect our code
    #we refere to our function as either a search which return the url for indicator list and their codes and 
    #return to return the data associated with the chosen indicator
    if action == "search":
        link = "https://ghoapi.azureedge.net/api/indicator?$filter=contains(IndicatorName,'%s')"
        return link
    
    elif action == "return":
        link = "https://ghoapi.azureedge.net/api/%s" 
        return link
    
    else: #to return None if the input text is not 'search' or 'return'
        return None

#Here I define a function to send a request to an url and get the response back and convert it to JSON if connection is made.
def request(link):
    
    try:
        response = rq.get(link) # Performing the get request.
        response.raise_for_status() #getting the success code
        obj = response.json() # converting the results to JSON format
        return [response.status_code, obj] #returning both the success code and the JSON object.
    
    except exceptions.HTTPError as httperr: #if something wrote with URL
        print("HTTP Error: ", httperr, "\nPossibily that indicator name does not exist, please visit https://ghoapi.azureedge.net/api/Indicator.")
        return None
    except exceptions.ConnectionError as connecterr : #if something wrong while connecting
        print("Connection Error: ", connecterr)
        return None
    except exceptions.Timeout  as timeouterr : #if connection time out
        print("Timeout Error: ", timeouterr)
        return None
    except exceptions.RequestException  as requesterr : #if any other thing happens
        print("Something went wrong ", requesterr)
        return None
        
#Here we define a function to convert JSON object to Pandas Dataframe.        
def json_df(json_object):
    obj = json_object['value'] #we only want to return the nested data under "value"
    if len(obj) ==0: #if in case connection was made but return was emtpy to return None
        return None
    else:
        df = pd.json_normalize(obj) #convert JSON to dataframe if the JSON have content
        return df

#Here we define a function to return the available indicators based on keywords
#The function takes either one value or an array of values
def find_indicator(param):
    
    if isinstance(param, str) or isinstance(param, int): #check if the entry datatype is string or integer
        link = link_generator("search") %param #generating a link using the link_generator function and the given keyword
        obj = request(link) #Make a request to the generated link using request function we defined earlier
        df = json_df(obj[1]) #convert the JSON Object to Pandas Dataframe
        if df is None: #if for whatever reason the dataframe comes empty to return None and give a message
            print("perhaps choose a different keyword")
            return None
        else:
            return df       
        
    elif isinstance(param, list): #check if the entry datatype is list
        json = [] #defining an empty array to input all of the indicators from keywords
        for i in param: # for loop to loop through the list of arrays
            link = link_generator("search") %i #generating a link using the link_generator function and the given keyword
            obj = request(link) #Make a request to the generated link using request function we defined earlier
            df_s = json_df(obj[1]) #convert the JSON Object to Pandas Dataframe
            json.append(df_s) #Collecting all the JSON objects together
        df = pd.concat(json) #convert all the JSON into one pandas datafarme
        if df is None: #if for whatever reason the dataframe comes empty to return None and give a messag
            print("perhaps choose a different keywords")
            return None
        else:
            return df
            
    else: #if any other datatype is passed on to return None
        print('Illigal data type was entered')
        return None
        
#Here we return the data associated with the 
def return_indicator(param):
    link = link_generator("return") %param
    obj = request(link)
    if obj is None:
         return None
    else:
        df = json_df(obj[1])
        if df is None:
            print("perhaps choose a different keywords")
            return None
        else:
            return df
        
```

```python
# This function will return a dataframe with all column names with their count of null values and their data types
def nulls_types(df):
    try:
        nulls = df.isnull().sum()
        types = df.dtypes
        dataset = pd.DataFrame({'Nulls': nulls, 'dtypes': list(types)})
        print('Total Row:', df.shape[0])
        return dataset
    except pd.errors.EmptyDataError:
        print("the input can only be Pandas Data frame")
    except AttributeError:
        print("This function only take pandas data fram please ensure to pass only pandas dataframe")
        
#Define a function to go through the data frame and change the data type for each column based on a python dictionary 
# key:columns value:new dtypes
def change_type(df,types_dict):

        for i in df.columns:
            for a,b in types_dict.items():
                if i == a:
                    df[i]=df[i].astype(b)
        print("Data type change was successful")
        return df

#a function to detect non numeric value in a column
def non_num(df,col):
    for i in df[col]:
        try:
            float(i)
        except ValueError:
            print ("None number value found: ", i)
    print("Process Succeed, not found")
    
    
#a function if we want to remove a specific value from our dataframe preferably 0 Null, None,NaN value
def replace_nan(df,col,value):
    vals = df.loc[df[col]== value, col].count()
    df[col].replace(value, np.nan,inplace=True)
    df.dropna(subset=[col],inplace=True)
    print("Changed: ", vals)
```

<h3 style="font-family:Times New Roman Bold">4.3.3 Outlier Removal</h3><br>
<font size="3"><div style="font-family:Times New Roman;text-align: justify">
As it can be seen that our Cholera cases data is quite skewed based on the desriptive statistic, however, since Cholera outbreaks in general have different distribution, this could be because of that, hence, we will remove the outliers and use the linear regression and correlation for both to see if it makes a difference.
</div></font>

```python
#a function remove outliers from our dataframe
#https://www.itl.nist.gov/div898/handbook/prc/section1/prc16.htm
def remove_outlier(df_in, col_name):
    count_1 = df_in[col_name].count()
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    df_out = df_in.loc[(df_in[col_name] > fence_low) & (df_in[col_name] < fence_high)]
    count_2 = df_out[col_name].count()
    print("Affected rows: ", count_1-count_2)
    return df_out
```

Regulatorization is implemented by introducing an Alpha parameter that can be used to draw the parameters closer to the center points. The first step in implementation regression analysis is to prepare the data, then choose the best regression model, then hypertuning the best predictive model, and finally evaluation of the model based and extract the feature importance.

<h3 style="font-family:Times New Roman Bold">6.1 Intial Preparations</h3><br>
<font size="3"><div style="font-family:Times New Roman;text-align: justify">
In regression analysis, the data must be separated into two slices: a dependent slice which determines which includes the parameters that are needed for prediction and an independent slice which determines which parameters need prediction. From this step forward, the dependent slice will be called 'x' and the independent slice 'y'.
After the x and y datasets have been constructed, both datasets are divided into two chunks, one for training the model and one for testing its accuracy.<br>
Lastly, the data for the parameters must be normalized since the data for the parameters are at different scales, such as percentages of population and climate, and thus, each parameter (column) requires normalization.

<h4 style="font-family:Times New Roman Bold">6.1.1 Builidng X,Y and Splitting the Dataset</h4><br>

The model will then be used to predict the test data and evaluated using regression standard evaluators such as mean absolute error (MAE), real and error mean ratios, R-squared and Root Mean Square Error (RMSE).

```python
#defining the model
rig_model = Ridge()
#defining the possible alpha parameters
parameter = {'alpha' : [1e-15, 1e-10,1e-8,1e-3 ,1e-2,1 ,5 ,10 ,20 ,30 ,35,40, 45, 50,55,100 ]}
#conducting the grid search to test out all the alpha parameters and seperate the data into 5 folds "cv = number of folds"
rigid_reg = GridSearchCV(rig_model, parameter, scoring='r2', cv=5 )
#traning the model
rigid_reg.fit(x_train, y_train)

#printing the best model parameters
print(rigid_reg.best_params_)
print(rigid_reg.best_score_)
```

<h3 style="font-family:Times New Roman Bold">2.4 Zonal Statistics</h3><br>
<font size="3"><div style="font-family:Times New Roman;text-align: justify">
Zonal Statistics is a method of calculating statistics for specific areas by using groupings. Observing the NetCDF file from section 2.2, the values are nested under a variable "timeseries-tas-annual-mean" for temperatures and "timeseries-pr-annual-mean" for precipitations. Consequently, these values must be extracted initially before proceeding with the process.<br>
As part of the zonal statistics, we need to pass a few arguments: first, the geometry for each country, then the value per specific year, then an affine to identify the boundaries of each geometry, since the rasterstats zonal statistics cannot read geometry directly from the dataset, therefore, the data with only variables must be exported and then imported back, using Rasterio affine function to generate the zonal statistics affine. Results will be provided in a Python dictionary that will need to be transformed into a pandas dataframe before being exported to CSV for use in this paper.

```python
#defining a function to convert dataframe to CSV
def exporting_csv(key,value):
    if not isinstance(key, str):
        print("Key must be string")
        return None
    
    if not isinstance(value, pd.DataFrame):
        print("value must be pandas dataframe")
        return None    
    
    if not os.path.exists('./climate/'): #if climate folder doesnot exist to create one
        os.makedirs('./climate/')
    
    if os.path.exists('./climate/'+key+"_final.csv") == True: #check if the file already exist
        os.remove('./climate/'+key+"_final.csv")              #delete the already present one
        print("file was found and was delected")
    
    return value.to_csv('./climate/'+key+"_final.csv", sep=','), print(key,": New file was created") #please change the directory if not applicable

#defining a function to export the xarray variables to a seperate nc file
def var_to_nc(var, name):
    if not isinstance(name, str):
        print("only string is allowed for name")
        return None
    try:
        if var.to_dataframe().empty == True:
            print("The data array is empty")
            return None
        else:
             if os.path.exists('./climate/'+name+'_mean.nc') == True:
                os.remove('./climate/'+name+'_mean.nc')
                print("file already existed and was delected.")
        return var.to_netcdf('./climate/'+name+'_mean.nc'), print(name,": New file was created with name: ", './climate/'+name+'_mean.nc') #please change the directory if not applicable
        
    except Exception as error:
        print("Possibly you tried a none data array, futher info: ", error)
        return None
```

<h4 style="font-family:Times New Roman Bold">Reading The exported datasets as Affine using Rasterio</h4><br>

```python
affine_precip = rio.open('./climate/precip_mean.nc').transform
affine_temp = rio.open('./climate/temp_mean.nc').transform

#define the study timeframe which is from 2000 to 2016
timeframe = list(range(2000,2017,1))
```

<h4 style="font-family:Times New Roman Bold">Generation of Mean Temperature Dataframe</h4><br>

```python
# go through all years
temp_dic = []
for year in timeframe:
    # get values of variable pear year
    nc_arr = temp_var.sel(year=year)
    nc_arr_vals = nc_arr.values
    # go through all geometries and compute zonal statistics
    for i in range(len(final_shp)):
        mean_mean=rstats.zonal_stats(final_shp.iloc[i, 1], nc_arr_vals, affine= affine_temp, stats="mean",all_touched=True, nodata=-9.96921e+36)
        result = [dict(item, country=final_shp.iloc[i, 0]) for item in mean_mean]
        result = [dict(item, elem=year) for item in result]
        temp_dic.append(result)
        
assert len(temp_dic) > 0, "process failed the dictionary is empty"

temp_df = None
for i in temp_dic:
    if temp_df is None:
        temp_df = pd.DataFrame.from_dict(i)
    else:
        pd1 = pd.DataFrame.from_dict(i)
        temp_df = pd.concat([temp_df,pd1])

assert isinstance(temp_df, pd.DataFrame), "temperature dictionary was not created"

#define dictionary
names = {'mean': 'Value',
             'country': 'SpatialDim',
             'elem': 'TimeDim'}

temp_df['IndicatorCode'] = "Temperature"
temp_df['SpatialDimType'] = 'COUNTRY'
temp_df['Comments'] = None 
#rename columns in DataFrame using dictionary
temp_df.rename(columns=names, inplace=True)
temp_df.head(3)
```
